```python
def LoanContact(LoandAwardID):
    found=False
    LoanContact_data = json.dumps({}, indent=4)
    col=[]
    with open(DataFile, newline='') as file_contents:
        for line in file_contents:
            key_value = line.strip().split(':')
            key = key_value[0].strip()  
            if (' Amount' in key) | (' Total' in key) | ('Total ' in key) | (' Balance' in key):
                value = key_value[1].strip().replace('$','').replace(',','')
            else:    
                value = key_value[1].strip()
            if found==False:
                if (key=='Loan Award ID') & (value==LoandAwardID):
                    found=True
            else:
                if LoanContactDDL.count([key])>=1:
                    if key == 'Loan Contact Type': 
                        Dict ={}
                        Dict[key] = value
                    elif key == 'Loan Special Contact':
                        Dict[key] = value
                        col.append(Dict)
                    else:
                        Dict[key] = value
                elif key == 'Loan Type Code':
                    break
    return col            
```

```python
def LoanStaus(LoandAwardID):
    found=False
    LoanStaus_data = json.dumps({}, indent=4)
    col=[]
    with open(DataFile, newline='') as file_contents:
        for line in file_contents:
            key_value = line.strip().split(':')
            key = key_value[0].strip()  
            if (' Amount' in key) | (' Total' in key) | ('Total ' in key) | (' Balance' in key):
                value = key_value[1].strip().replace('$','').replace(',','')
            else:    
                value = key_value[1].strip()
            if found==False:
                if (key=='Loan Award ID') & (value==LoandAwardID):
                    found=True
            else:
                if LoanStausDDL.count([key])>=1:
                    if key == 'Loan Status': 
                        Dict ={}
                        Dict[key] = value
                    elif key == 'Loan Status Effective Date':
                        Dict[key] = value
                        col.append(Dict)
                    else:
                        Dict[key] = value
                elif key == 'Loan Disbursement Date':
                    break
    return col            

```

```python
# Get all Loand Detail data in Json Obejct 
data = {}
LoanDetail_data = json.dumps(data, indent=4)
col=[]

# Loop through the file contents and extract the key-value pairs
with open(DataFile, newline='') as file_contents:
    for line in file_contents:
        # Split each line by the colon character
        key_value = line.strip().split(':')
        # Use the first element as the key and the second element as the value
        key = key_value[0].strip()  #.replace(" ", "_")
        if (' Amount' in key) | (' Total' in key) | ('Total ' in key) | (' Balance' in key):
            value = key_value[1].strip().replace('$','').replace(',','')
        else:    
            value = key_value[1].strip()
        if LoanDetailDDL.count([key])>=1:
            if key == 'Loan Type Code':                
                Dict ={}
                #LSkey=0
                Dict[key] = value
            elif key == 'Loan Award ID':
                Dict[key] = value
                LoandAwardID=value
            elif key == 'Loan Disbursement Amount':
                Dict[key] = value
                Dict['LoanSatus'] = LoanStaus(LoandAwardID)
                Dict['LoanContact'] = LoanContact(LoandAwardID)
                col.append(Dict)
                #LSCol=[]
            else:
                Dict[key] = value
LoanDetail_data = json.dumps(col, indent=4)
```
